```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# we can download dateset from kaggle
# https://www.kaggle.com/datasets/shayanfazeli/heartbeat

# REF:
# https://arxiv.org/pdf/1805.00794.pdf
```

```python
## Load data
df_train = pd.read_csv("mitbih_train.csv", header=None)
df_test = pd.read_csv("mitbih_test.csv", header=None)

train_X = df_train.values[:, :-1]
train_y = df_train.values[:, -1].astype(int)
test_X = df_test.values[:, :-1]
test_y = df_test.values[:, -1].astype(int)
plt.hist(train_y)
plt.xlabel("Labels")
plt.ylabel("The numbers of labels")
plt.show()
```

```python
## Get each label
C0 = np.argwhere(train_y == 0).flatten() # Cat.N
C1 = np.argwhere(train_y == 1).flatten() # Cat.S
C2 = np.argwhere(train_y == 2).flatten() # Cat.V
C3 = np.argwhere(train_y == 3).flatten() # Cat.F
C4 = np.argwhere(train_y == 4).flatten() # Cat.Q
```

image.png

```python
x = np.arange(0, 187)*1/125 #125[Hz]

plt.figure(figsize=(10,6))
plt.plot(x, train_X[C0, :][0], label='Cat.N')
plt.plot(x, train_X[C1, :][0], label='Cat.S')
plt.plot(x, train_X[C2, :][0], label='Cat.V')
plt.plot(x, train_X[C3, :][0], label='Cat.F')
plt.plot(x, train_X[C4, :][0], label='Cat.Q')
plt.title('1-beat ECG for every category', fontsize=20)
plt.ylabel('ECG values', fontsize=15)
plt.xlabel('Time[ms]', fontsize=15)
plt.grid(True), plt.legend(), plt.show()
```

```python
x = np.arange(0, 187)*1/125 #125[Hz]
plot_num = 50
labels = [C0, C1, C2, C3, C4]
labels_str = ["Cat.N", "Cat.S", "Cat.V", "Cat.F", "Cat.Q"]

for i, label in enumerate(labels_str):
    plt.figure(figsize=(10,6))
    for n in range(plot_num):
        plt.plot(x, train_X[labels[i], :][n], alpha=0.25)
    plt.plot(x, np.mean(train_X[labels[i], :], axis=0), label='Mean Value', linewidth=3)
    plt.title(('1-beat ECG for ' + label), fontsize=20)
    plt.ylabel('ECG values', fontsize=15)
    plt.xlabel('Time[ms]', fontsize=15)
    plt.grid(True)
    plt.legend(), plt.show()
```
